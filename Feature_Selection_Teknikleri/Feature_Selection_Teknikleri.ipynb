```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd

df = pd.read_csv(r"C:\Dataset\SQLiV3.csv")
df.drop('Unnamed: 2', axis=1, inplace=True)
df.drop('Unnamed: 3', axis=1, inplace=True)

df=df.dropna()
df.head()
```

```python
df['Sentence'] = pd.to_numeric(df['Sentence'], errors='coerce').fillna(0, downcast="float")

X = df.drop(columns=['Label'])
y = df['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state = 0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)

X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.decomposition import NMF
NMF = NMF(n_components=2, init='random', random_state=0)
X_train = NMF.fit_transform(X_train)
X_test = NMF.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
LDA=LDA.fit(X_train, y_train)
y_pred = LDA.predict(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
df = pd.read_csv("C:\Dataset\CM1.csv")
df.head()
```

```python
X = df.drop(columns=['Defective'], axis=1)
y = df['Defective']
```

```python
X.shape
```

```python
X.head()
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.svm import SVC

SVC = SVC(gamma='auto')

sfs1 = sfs(SVC, k_features=5, forward=False, verbose=1, scoring='accuracy')
```

```python
sfs1 = sfs1.fit(X, y)
```

```python
feat_names = list(sfs1.k_feature_names_)
print(feat_names)
```

```python
new_df = df[feat_names]
new_df['Defective'] = df['Defective']

new_df.head()
```

```python
X = new_df.drop(columns=['Defective'], axis=1)
y = new_df['Defective']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state = 0)
```

```python
from sklearn.preprocessing import Normalizer

normalizer = Normalizer()
X_train = normalizer.fit_transform(X_train)
X_test = normalizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=42).fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
