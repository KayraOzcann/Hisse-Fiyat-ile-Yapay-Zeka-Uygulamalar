```python
import pandas as pd
```

```python
dataset = pd.read_csv('C:\DataSet\Malware-Label.csv')

dataset.head()
```

```python
X = dataset.drop(columns=['Label'])
y = dataset['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,random_state = 0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()


from sklearn.neural_network import BernoulliRBM
rbm1 = BernoulliRBM(n_components=X_train.shape[1], learning_rate=0.06, n_iter=100, verbose=1,  random_state=101)

rbm = Pipeline(steps=[("rbm1", rbm1), ("log", log)])
rbm.fit(X_train.values, y_train)
y_pred = rbm.predict(X_test.values)
```

```python
from sklearn import metrics
print("doÄŸrulama:",metrics.accuracy_score(y_test,y_pred))
```
